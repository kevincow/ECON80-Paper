```python
df_growth = pd.read_csv('./gdp_growth.csv')
df_growth = df_growth.rename(columns={'date': 'Year'})
df_growth['Year'] = df_growth['Year'].apply(lambda x: int(str(x)[:4]))
df_growth
```

```python
df_productivity = pd.read_csv('./productivity_growth.csv')
df_productivity = df_productivity.rename(columns={'date': 'Year'})
df_productivity['Year'] = df_productivity['Year'].apply(lambda x: int(str(x)[:4]))
df_productivity
```

```python
df_min_wage = pd.read_csv('min_wage_pre2020.csv')
df_min_wage = df_min_wage[['Year', 'State', 'Effective.Minimum.Wage']]
df_min_wage = df_min_wage.rename(columns={
    'Year': 'year',
    'State': 'state', 
    'Effective.Minimum.Wage': 'min_wage'
})

states = df_min_wage['state'].unique()
years = [2021, 2022, 2023]

new_data = pd.DataFrame([(year, state, 7.25) for state in states for year in years],
                        columns=['year', 'state', 'min_wage'])

df_min_wage = pd.concat([df_min_wage, new_data], ignore_index=True)
df_min_wage.tail(50)
```

```python
df_post_min_wage = pd.read_csv('./min_wage_post2020.csv')
df_post_min_wage = df_post_min_wage.rename(columns={'DATE': 'year'})
df_post_min_wage['year'] = df_post_min_wage['year'].apply(lambda x: int(str(x)[:4]))
df_post_min_wage = df_post_min_wage.melt(id_vars='year', var_name='state', value_name='min_wage')
df_post_min_wage = df_post_min_wage[df_post_min_wage['year'] > 2020]
df_post_min_wage = df_post_min_wage.dropna()

df_min_wage = pd.merge(df_min_wage, df_post_min_wage, on=['year', 'state'], how='left')

df_min_wage['min_wage_x'] = df_min_wage.apply(
    lambda row: max(row['min_wage_x'], row['min_wage_y']) if row['year'] > 2021 else row['min_wage_x'],
    axis=1
)

df_min_wage = df_min_wage[['year', 'state', 'min_wage_x']]

df_min_wage = df_min_wage.sort_values(by=['state', 'year']).reset_index(drop=True)
df_min_wage['min_wage_increase'] = df_min_wage.groupby('state')['min_wage_x'].diff().fillna(0).apply(lambda x: 1 if x > 0 else 0)

df_min_wage = df_min_wage.rename(columns={'min_wage_x': 'min_wage', 'year': 'Year', 'state': 'State'})
df_min_wage = df_min_wage.sort_values(by=['Year', 'State'], ascending=True)
df_min_wage.tail(70)
```

```python
df_pop = pd.read_csv('population_size.csv')
df_pop = df_pop.iloc[80:]
df_pop = df_pop.reset_index(drop=True)
df_pop['Year'] = pd.to_datetime(df_pop['DATE'], format='%m/%d/%y')
df_pop['Year'] = df_pop['Year'].dt.year
df_pop = df_pop.drop(columns=df_pop.columns[0])
df_pop = df_pop.dropna()
df_pop = df_pop.melt(id_vars='Year', var_name='State', value_name='population')
df_pop
```
